### 1. **Сумма цифр числа**

Напишите программу, которая получит четырехзначное число от пользователя и выведет на экран сумму цифр
этого числа.

**Пример вывода:**

Введите четырехзначное число: 1634

Сумма цифр числа: 14





````python
number = int(input("Введите четырехзначное число: "))

digit1 = number // 1000 % 10  # 1634 -> 1.634 -> 1
digit2 = (number // 100) % 10  # 1634 -> 16.34 -> 16 -> 6
digit3 = (number // 10) % 10  # 1634 -> 163.4 -> 163 -> 3
digit4 = (number // 1) % 10  # 1634 -> 4

print(digit1, digit2, digit3, digit4)

sum_digits = digit1 + digit2 + digit3 + digit4
print("Сумма цифр числа: ", sum_digits)
````

умножает одно число на второе и выводит результат и оба числа через дефис.
Не сохраняйте результат умножения в переменной.

Введите первое число: 4

### 3. **Вычисление без операторов % и //**

Напишите программу, которая получает два числа от пользователя и вычисляет:  
- Остаток от деления
- Результат целочисленного деления

Решить без использования операторов `%` и `//`.

**Пример ввода:**
```
Введите первое число: 10  
Введите второе число: 3
```

**Пример вывода:**
```
Остаток от деления: 1  
Целочисленное деление: 3
```


````python
a = int(input("Введите первое число: "))
b = int(input("Введите второе число: "))

# Вычисление целочисленного деления
integer_division = int(a / b)

# Вычисление остатка
remainder = a - (integer_division * b)

print("Остаток от деления:", remainder)
print("Целочисленное деление:", integer_division)
````

### 3. **\* Стоимость мобильного тарифа**

Напишите программу для расчёта стоимости использования мобильного тарифа:
- Базовая стоимость: 30 евро.
- Каждый мегабайт интернета сверх 500 МБ стоит 0.2 евро.

Программа должна запросить у пользователя количество использованных мегабайтов и вывести сколько всего он заплатил (базовый и переплата).

**Пример вывода:**

Введите использованные мегабайты: 510  

Общая стоимость: 32.0


````python
base_cost = 30
base_mb = 500
extra_cost = 0.2

mb_used = int(input("Введите использованные мегабайты: "))

extra_mb_cost = (mb_used - 500) * extra_cost
total_cost = base_cost + (mb_used > base_mb) * extra_mb_cost

print("Общая стоимость:", total_cost)
````
